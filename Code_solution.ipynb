# Extracting Technical (Hard) skills from Raw_dataset

```python
def whether_capital(df):
    
    a = 0
    b = []
    for skill in df:
        a = 0
        for word in skill.split():
            if word[0].islower():
                a+=1
        if (a > 0):
            b.append(0)
        else:
            b.append(1)
        
    return b
```

```python
req = df['RAW DATA']
res = whether_capital(df['RAW DATA'])
for i in range(0,len(req)):
    print(req[i]," : ",res[i])
```

```python
new_res_df = df
new_res_df['Tech_Skills'] = res
```

```python
tech_skills = new_res_df[new_res_df['Tech_Skills'] == 1]['RAW DATA']
tech_skills
```
